Función que calcule $\beta_n$, verificar que se cumple para $\beta_2$ con una función.

```python
import numpy as np
from scipy.special import comb

def calcular_beta_n(m, alphas, n):
    beta_n = 0
    for k in range(n + 1):
        term = comb(n, k) * (-m)**(n - k) * alphas[k]
        beta_n += term
    return beta_n
```

Suponiendo una distribución normal con media m = 0 y momentos hasta α_2:

```python
m = 0
alphas = [1, 0, 1]  # α_0 = 1 (momento 0), α_1 = 0 (momento 1), α_2 = 1 (momento 2)
n = 2

# Calcular β_2
beta_2 = calcular_beta_n(m, alphas, n)
print(f"β_2: {beta_2}")

x = np.random.normal(m, 1, 1000)  # 1000 muestras de una distribución normal
varianza = np.var(x)
print(f"Varianza: {varianza}")
```
