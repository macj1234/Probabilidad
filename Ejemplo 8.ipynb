Definición: Coeficiente de correlación

Si $E[x_{1}^{2}]$ y $E[x_{2}^{2}]$ son finitos y las varianzas $\sigma_{1}^2$ y $\sigma_{2}^2$ respectivos son mayores de cero, entonces

$$ \rho(x_1, x_2) = \frac{Cov(x_1, x_2)}{\sigma_{1} \sigma_{2}}$$

Nota: Si $x_1$, $x_2$ son independientes, entonces $\rho(x_1, x_2) = 0$, lo contrario no siempre se cumple

Se generan las variables aleatorias $x_1$ y $x_2$

```python
import numpy as np

x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(0, 1, 1000)

# Calcular la covarianza entre X1 y X2
covarianza = np.cov(x1, x2)[0, 1]

# Calcular la desviación estándar de X1 y X2
sigma_x1 = np.std(x1)
sigma_x2 = np.std(x2)

# Calcular el coeficiente de correlación
if sigma_x1 > 0 and sigma_x2 > 0:
    correlacion = covarianza / (sigma_x1 * sigma_x2)
    print(f"ρ(x1, x2): {correlacion}")

```
