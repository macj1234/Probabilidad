Programa para calcular la distribución de probabilidad binomial para cualquier
(k, n) y graficar.

Calcula el coeficiente binomial $\begin{pmatrix} n \\ k \end{pmatrix}$.

```python
import matplotlib.pyplot as plt

def coeficiente_binomial(n, k):
    if k > n:
        return 0
    if k == 0 or k == n:
        return 1
    k = min(k, n - k)
    c = 1
    for i in range(k):
        c = c * (n - i) // (i + 1)
    return c
```

Calcula la función de masa de probabilidad de la distribución binomial.

```python
def binomial_pmf(k, n, p):
    return coeficiente_binomial(n, k) * (p ** k) * ((1 - p) ** (n - k))
```

Calcula la función de distribución acumulada de la distribución binomial.

```python
def binomial_cdf(k, n, p):
    return sum(binomial_pmf(i, n, p) for i in range(k + 1))
```
